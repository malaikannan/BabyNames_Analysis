```python
USA_baby_names_1976 = USA_babynames_dataframe[USA_babynames_dataframe['year']>1976]
USA_baby_names_1976_1994 = USA_baby_names_1976[USA_baby_names_1976['year']<1994]
total_births = USA_baby_names_1976_1994.pivot_table('count',index='state',columns='sex',aggfunc=sum)
total_births['Female'] = (total_births['F']/(total_births['F']+total_births['M']))*100
total_births['Male'] = (total_births['M']/(total_births['F']+total_births['M']))*100
total_births['proportion'] = total_births['Male']-total_births['Female']
total_births.sort(columns='proportion')


good_sex_ratio = total_births.sort(columns='proportion').head(5)
good_sex_ratio = total_births.sort(columns='proportion').head(5)
good_sex_ratio = good_sex_ratio.iloc[:,0]
good_sex_ratio = total_births.sort(columns='proportion').head(5)
good_sex_ratio = good_sex_ratio.iloc[:,0]


bad_sex_ratio = total_births.sort(columns='proportion').tail(5)
bad_sex_ratio = bad_sex_ratio.iloc[:,0]


```

Below function is plot good sexratio states vs bad sex ratio states

```python
#Below function is plot good sexratio states vs bad sex ratio states
def plot_states(good_sex_ratio_states,good_sex_ratio_color,bad_sex_ratio_states,bad_sex_ratio_color,lower_48,alaska,hawaii,gs):

    #Loading US State shape file
    if os.path.exists(os.path.join(basemap_data_dir,"states.shp")):
        shpf = shapefile.Reader(os.path.join(basemap_data_dir,"states"))
    shapes = shpf.shapes()
    records = shpf.records()

    for record, shape in zip(records,shapes):
        #read shape
        lons,lats = zip(*shape.points)
        #Checking whether state is Alaska
        if record[4] =='AK':
            data = np.array(alaska(lons, lats)).T
            ax = plt.subplot(gs[1])
        #Checking whether state is Hawaii
        elif record[4]=='HI':
            data = np.array(hawaii(lons, lats)).T
            ax = plt.subplot(gs[2])
        else:   
            data = np.array(lower_48(lons, lats)).T
            ax = plt.subplot(gs[0])
        #each shape may have different segments
        if len(shape.parts) == 1:
            segs = [data,]
        else:
            segs = []
            for i in range(1,len(shape.parts)):
                index = shape.parts[i-1]
                index2 = shape.parts[i]
                segs.append(data[index:index2])
            segs.append(data[index2:])

        #draws the segments, and sets its properties. A colormap is used to get the gradient effect.
        #checking to see whether the record is in the good sex ratio states
        if record[4] in good_sex_ratio_states:
            lines = LineCollection(segs,antialiaseds=(1,))
            #setting different color for good_sex_ratio states
            lines.set_facecolors(good_sex_ratio_color)
            lines.set_edgecolors('k')
            lines.set_linewidth(0.1)
            ax.add_collection(lines)
        elif record[4] in bad_sex_ratio_states:
            lines = LineCollection(segs,antialiaseds=(1,))
            #setting different color for bad_sex_ratio states
            lines.set_facecolors(bad_sex_ratio_color)
            lines.set_edgecolors('k')
            lines.set_linewidth(0.1)
            ax.add_collection(lines)
```

```python
import matplotlib.patches as mpatches


#Method to plot different colors for different state based on name distribution. 
#Needs State and name in state list and color to be used for in the color list

def plot_states_name_distribution(state_list,color_list,lower_48,alaska,hawaii,gs):


    #Opening US States shape file
    if os.path.exists(os.path.join(basemap_data_dir,"states.shp")):
        shpf = shapefile.Reader(os.path.join(basemap_data_dir,"states"))
    
    shapes = shpf.shapes()
    records = shpf.records()
    color_patches = {}

    #iterating through records in the shape file.
    for record, shape in zip(records,shapes):
        #read shape
        
        lons,lats = zip(*shape.points)
        #Checking for whether the state is alaska
        if record[4] =='AK':
            data = np.array(alaska(lons, lats)).T
            ax = plt.subplot(gs[1])
        #Checking for whether the state is Hawaii
        elif record[4]=='HI':
            data = np.array(hawaii(lons, lats)).T
            ax = plt.subplot(gs[2])
        else:   
            data = np.array(lower_48(lons, lats)).T
            ax = plt.subplot(gs[0])
        #each shape may have different segments
        if len(shape.parts) == 1:
            segs = [data,]
        else:
            segs = []
            for i in range(1,len(shape.parts)):
                index = shape.parts[i-1]
                index2 = shape.parts[i]
                segs.append(data[index:index2])
            segs.append(data[index2:])
        
        #Retreiving the popular baby name from State_list dictionary for the current state
        name = state_list[record[4]]
        #Retreiving the color to be used for the corresponding name.
        color = color_list[name]
        lines = LineCollection(segs,antialiaseds=(1,))
        lines.set_facecolors(color)
        lines.set_label(name)
        lines.set_edgecolors('k')
        lines.set_linewidth(0.1)
        ax.add_collection(lines)
       
        #Adding a legend for the name if the name does not exist in the color_patches dictionary    
        if color in color_patches:
            color_patches = color_patches
        else:
            color_patches[color] = mpatches.Patch(color=color, label=name)
            
    legend_patches = []
    rownum = 0
    
    #Iterating through color_patches and adding Patch object to legend_patches dictionary.
    for key in color_patches:
          legend_patches.append(color_patches[key])
    
    #Binding Legend Patches to bbox_to_anchor .
    plt.legend(handles=legend_patches,bbox_to_anchor=(2.50, 0.25),loc=3)
    plt.show()

            
```

```python
# Method to find most popular names by years
def popular_names_by_years(from_year,to_year,sex):
    
    popular_name_by_years = USA_babynames_dataframe[(USA_babynames_dataframe['year']>=from_year)&(USA_babynames_dataframe['year']<=to_year)]
    popular_name_by_years = popular_name_by_years.drop('year',1)
    
    #checking to whether M or F is passed as value for Sex. If so filter the data frame on Sex column
    if sex !='both':
        popular_name_by_years = popular_name_by_years[popular_name_by_years['sex']==sex]
        
    
    popular_name_by_years = popular_name_by_years.drop('sex',1)
    popular_name_by_years = popular_name_by_years.groupby(['state','name'],sort=True).sum()
    popular_name_by_years = popular_name_by_years.sort(['count'], ascending=False)
    
    #Slicing and Re-indexing
    popular_name_by_years = pd.DataFrame(popular_name_by_years.iloc[:,0])
    popular_name_by_years = popular_name_by_years.reset_index()
    
    popular_name_by_years_dict = {}
    for index,row in popular_name_by_years.iterrows():
        if(row[0] in popular_name_by_years_dict):
            row[2] = row[2]
        else:
            popular_name_by_years_dict[row[0]] = row[1]

    return popular_name_by_years_dict
```

```python
USA_babynames_without_states = USA_babynames_dataframe[['sex','year','name','count']]

def to_number(s):
    try:
        s1 = float(s)
        return s1
    except ValueError:
        return s

def add_prop(group):
    # To get the value in float for calculating proportion
    births = group.apply(lambda group : to_number(group['count']) , axis = 1)

    group['prop'] = births / births.sum()
    return group


USA_babynames_without_states = USA_babynames_without_states.groupby(['year', 'sex']).apply(add_prop)
```

```python
USA_babynames_without_states = USA_babynames_without_states.sort(columns='prop',ascending=False)

USA_babynames_without_states.head()
```

#### Babies named after Hurricanes?    Whenever there is a Category 3 hurricane or higher strikes then kids have been named after the hurricane or similar sounding name  I referred to the list of hurricanes in this link http://www.wunderground.com/hurricane/retired.asp.  Hurrican Ivan and Isabel seem to have had an impact with naming babies. There is a spike in the baby names sounding like Ivan or Isable after ther hurricane hit.    Below Soundex method is to convert name to a soundex function

```python
from itertools import groupby
 
def soundex(word):
   codes = ("bfpv","cgjkqsxz", "dt", "l", "mn", "r")
   soundDict = dict((ch, str(ix+1)) for ix,cod in enumerate(codes) for ch in cod)
   cmap2 = lambda kar: soundDict.get(kar, '9')
   sdx =  ''.join(cmap2(kar) for kar in word.lower())
   sdx2 = word[0].upper() + ''.join(k for k,g in list(groupby(sdx))[1:] if k!='9')
   sdx3 = sdx2[0:4].ljust(4,'0')
   return sdx3

```

#### Below Analysis is for finding gender ambiguous name.   Since i need to calculate gender ambigous name for 1945 and 2013. I created the below method.

```python
# calculate set of male_only, female_only, ambigender names
import pandas as pd

#Implemented set operation to find the ambigous name in Male and Female sets. 
def ambi_sex_names(year):
    
    USA_babynames_dataframe_year = USA_babynames_dataframe[USA_babynames_dataframe['year']==year]
    k = USA_babynames_dataframe_year.groupby('sex').apply(lambda s: set(list(s['name'])))
    male_only_names = k['M'] - k['F']
    female_only_names = k['F'] - k['M']
    ambi_names = k['F'] & k['M'] # intersection of two 
    return {'male_only_names': male_only_names, 
            'female_only_names': female_only_names,
            'ambi_names': ambi_names }
    
def calc_of_sex_of_names(year):
    
    USA_babynames_filteredby_year = USA_babynames_dataframe[USA_babynames_dataframe['year']==year]
    USA_babynames_filteredby_year = USA_babynames_filteredby_year.drop(['year','state','sex'],1)
    USA_babynames_filteredby_year = USA_babynames_filteredby_year.groupby('name').sum()
    USA_babynames_filteredby_year = USA_babynames_filteredby_year.sort('count',ascending = False)
    USA_babynames_filteredby_year = pd.DataFrame(USA_babynames_filteredby_year.iloc[:,0])
    USA_babynames_filteredby_year = USA_babynames_filteredby_year.reset_index()
    names_by_sex = ambi_sex_names(year) 
    ambi_names_dataframe = pd.DataFrame(list(names_by_sex['ambi_names']),columns=['name'])
    ambi_names_dataframe_merged = ambi_names_dataframe.merge(USA_babynames_filteredby_year,on='name',sort=True)
    ambi_names_dataframe_return = ambi_names_dataframe_merged.sort('count',ascending = False).head(1)
    
    return ambi_names_dataframe_return
```

```python
def get_quantile(group,quantile=0.5):
    df = group.sort_index(by='prop',ascending=False)
    df = pd.DataFrame(df.prop.cumsum().searchsorted(quantile))
    return df
```

#### Below analysis is to find largest increase and decrease in popularity in names 

```python
USA_babynames_without_states = USA_babynames_without_states.sort(columns='prop',ascending=False)


def Percentage_Increase_Decrease(from_year,to_year):
    names_from_year = USA_babynames_without_states[USA_babynames_without_states.year==from_year]
    names_from_year = names_from_year.drop(['sex','year','count'],axis=1)
    names_from_year = names_from_year.groupby('name').sum()
    names_from_year = pd.DataFrame(names_from_year.iloc[:,0])
    names_from_year = names_from_year.reset_index()
    names_rank_year_merged = names_from_year
    
    for year in range(from_year,to_year):
        names_next_year = USA_babynames_without_states[USA_babynames_without_states.year==year]
        names_next_year = names_next_year.drop(['sex','year','count'],axis=1)
        names_next_year = names_next_year.groupby('name').sum()
        names_next_year = pd.DataFrame(names_next_year.iloc[:,0])
        names_next_year = names_next_year.reset_index()
        merge_left_suffix = "_" + str((year))
        merge_right_suffix = "_" + str(year+1)
        left_column_name = "prop"+merge_left_suffix
        right_column_name = "prop"+merge_right_suffix
        names_rank_year_merged = names_from_year.merge(names_next_year,on='name',suffixes = (merge_left_suffix,merge_right_suffix),sort=True)
        
        names_rank_year_merged['diff'] = names_rank_year_merged[left_column_name] - names_rank_year_merged[right_column_name]
        if year==from_year:
            names_rank_year_merged['diff'] = names_rank_year_merged[left_column_name] - names_rank_year_merged[right_column_name]
        else:
            names_rank_dummy = names_rank_year_merged
            names_rank_year_merged['diff'] = names_rank_dummy['diff'] + names_rank_year_merged[left_column_name] - names_rank_year_merged[right_column_name]
            
        names_from_year = names_next_year
    return names_rank_year_merged



```

```python
df_1980 = Percentage_Increase_Decrease(1980,2013)
df_1980 = df_1980.sort('diff',ascending = False)
#Largest Decrease in popularity from 1980
df_1980.head(5)
```
